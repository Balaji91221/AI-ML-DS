```python
np.random.seed(23)
df = pd.DataFrame(np.random.randn(4,5), index = ['a', 'b', 'c', 'd'], columns = ['x', 'y', 'z', '1', '2'])
df
```

```python
# Extract negative values

df.values[df.values < 0]
```

```python
df.values
```

```python
df.values < 0
```

```python
# With index

for c in df.columns:
    for i in df.index:
        if df[c][i] < 0:
            print(c, i, df[c][i])
```

```python
# With index

for c in df.columns:
    for i in df.index:
        if df[c][i] < 0:
            print(c, i, df.loc[i, c])
```
