### Subplots

```python
x=np.linspace(1,15,10)
y=x**2
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'b--')
plt.subplot(2,2,2)
plt.plot(x,y,'r*')
plt.subplot(2,2,3)
plt.plot(x,y,'g*')
plt.subplot(2,2,4)
plt.plot(x,y,'r>')
```

```python
fig=plt.figure()
ax1=fig.add_axes([0.1,0.1,0.8,0.8])
ax2=fig.add_axes([0.2,0.5,0.4,0.3])

ax1.plot(x,y)

ax2.plot(y,x)
```

### Pie Charts

```python
import matplotlib.pyplot as plt
labels=["php","pyhton","java","c"]
numbers=[75,70,80,60]
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])  #[left,bottom,width,height]
axes1.pie(numbers,labels=labels,autopct="%0.2f%%",colors=["orange","pink","green","blue"])
axes1.legend()
```

```python
#=["php","pyhton","java","c"]
Explode=[0.5,0,0,0.2]
plt.pie(numbers,labels=labels,explode=Explode)
```
